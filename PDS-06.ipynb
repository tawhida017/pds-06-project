```python
# Count plot of cardio data

f, ax=plt.subplots(1,2,figsize=(10,5))

# Get the unique values in the 'SSC_GPA' column
unique_gpa_values = df['SSC_GPA'].unique()

# Create an explode list with the same length as the unique GPA values
# This will explode all slices by a small amount (0.1)
explode = [0.1] * len(unique_gpa_values)

# Replace 'GPA' with the actual column name in your DataFrame containing GPA data
# Assuming the correct column name is 'SSC_GPA' or 'HSC_GPA' based on your other code snippets
# For example, if the column name is 'SSC_GPA':
df['SSC_GPA'].value_counts().plot.pie(explode=explode, autopct='%1.1f%%', ax=ax[0],shadow=True)
ax[0].set_title('SSC_GPA') # Update the title accordingly
ax[0].set_ylabel('')

# Pass the DataFrame 'df' to the 'data' argument in countplot
sb.countplot(x='SSC_GPA', data=df, ax=ax[1])  # Add data=df

ax[1].set_title('SSC_GPA') # Update the title accordingly

# Get the counts of all unique values in 'SSC_GPA'
counts = df['SSC_GPA'].value_counts()

# Print the counts for each unique GPA value
for gpa, count in counts.items():
    print(f"SSC_GPA {gpa}: {count}")

plt.grid()
plt.show()
```
